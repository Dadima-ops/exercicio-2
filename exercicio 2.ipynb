```python
# exercicio08
nome = input("Nome: ")
produto = input("Produto: ")
quantidade = int(input("Quantidade: "))
preco = float(input("Preço: "))
total = quantidade * preco
print(f"Resumo da compra:")
print(f"Nome: {nome}")
print(f"Produto: {produto}")
print(f"Quantidade: {quantidade}")
print(f"Preço unitário: R${preco:.2f}")
print(f"Valor total: R${total:.2f}")
```

```python
# exercicio09
numero = int(input("Número: "))

# Multiplicação
print(f"{numero} x 1 = {numero * 1}")
print(f"{numero} x 2 = {numero * 2}")
print(f"{numero} x 3 = {numero * 3}")
print(f"{numero} x 4 = {numero * 4}")
print(f"{numero} x 5 = {numero * 5}")
print(f"{numero} x 6 = {numero * 6}")
print(f"{numero} x 7 = {numero * 7}")
print(f"{numero} x 8 = {numero * 8}")
print(f"{numero} x 9 = {numero * 9}")
print(f"{numero} x 10 = {numero * 10}")
print()

# Divisão
print(f"{numero} ÷ 1 = {numero / 1:.2f}")
print(f"{numero} ÷ 2 = {numero / 2:.2f}")
print(f"{numero} ÷ 3 = {numero / 3:.2f}")
print(f"{numero} ÷ 4 = {numero / 4:.2f}")
print(f"{numero} ÷ 5 = {numero / 5:.2f}")
print(f"{numero} ÷ 6 = {numero / 6:.2f}")
print(f"{numero} ÷ 7 = {numero / 7:.2f}")
print(f"{numero} ÷ 8 = {numero / 8:.2f}")
print(f"{numero} ÷ 9 = {numero / 9:.2f}")
print(f"{numero} ÷ 10 = {numero / 10:.2f}")
print()

# Adição
print(f"{numero} + 1 = {numero + 1}")
print(f"{numero} + 2 = {numero + 2}")
print(f"{numero} + 3 = {numero + 3}")
print(f"{numero} + 4 = {numero + 4}")
print(f"{numero} + 5 = {numero + 5}")
print(f"{numero} + 6 = {numero + 6}")
print(f"{numero} + 7 = {numero + 7}")
print(f"{numero} + 8 = {numero + 8}")
print(f"{numero} + 9 = {numero + 9}")
print(f"{numero} + 10 = {numero + 10}")
print()

# Subtração
print(f"{numero} - 1 = {numero - 1}")
print(f"{numero} - 2 = {numero - 2}")
print(f"{numero} - 3 = {numero - 3}")
print(f"{numero} - 4 = {numero - 4}")
print(f"{numero} - 5 = {numero - 5}")
print(f"{numero} - 6 = {numero - 6}")
print(f"{numero} - 7 = {numero - 7}")
print(f"{numero} - 8 = {numero - 8}")
print(f"{numero} - 9 = {numero - 9}")
print(f"{numero} - 10 = {numero - 10}")

```
